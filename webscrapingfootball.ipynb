```python
from bs4 import BeautifulSoup
import requests
import random
import pandas as pd
```

```python
url = "https://fbref.com/en/comps/8/stats/Champions-League-Stats"
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
print(page)
```

```python
team_names_th = soup.find_all("th",{'class':'left','data-stat' : 'team'})
team_names = [(i.find("a")).text for i in team_names_th if str(i.find("a").text)[:2] != 'vs']
team_country = [(i.find("span")["title"]) for i in team_names_th if str(i.find("a").text)[:2] != 'vs']


dic = {}
table1 = soup.find("table",{"id":"stats_squads_standard_for"})
tbody = table1.find("tbody")
rows = tbody.find_all("tr")
for i in rows[0].find_all("td"):
    dic[i["data-stat"]] = []
dic["Squad"] = team_names


for row in rows:
    datam = row.find_all("td")
    for data in datam:
        dic[data["data-stat"]].append(float(data.text))
dic["Country"] = team_country
dfteams = pd.DataFrame.from_dict(dic)




filename = "Users/hajiaga/Desktop/DataScience/teams.csv"
import os

if os.path.exists(filename):
    os.remove(filename)
dfteams.to_csv(filename, index=False) 

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

path = "//Users//hajiaga//Desktop//DataScience//chromedriver"

driver = webdriver.Chrome()

driver.implicitly_wait(5)

driver.get("https://fbref.com/en/comps/8/stats/Champions-League-Stats")

time.sleep(20)

driver.find_element(By.XPATH , "//input[@type='checkbox' and @id = 'fs_check_stats_standard']")

content = driver.find_element(By.CSS_SELECTOR, "div[id*='div_stats_standard'")

lazimlilar = content.find_elements(By.TAG_NAME , "tr")
dicplayer = {}

for lazimli in lazimlilar:
    data_players = lazimli.find_elements(By.TAG_NAME , "td")
    for datax in data_players:
        if datax.get_attribute('data-stat') in dicplayer:
            dicplayer[datax.get_attribute('data-stat')].append(datax.text) 
        else:
            dicplayer[datax.get_attribute('data-stat')]= []
            dicplayer[datax.get_attribute('data-stat')].append(datax.text) 

driver.quit()
```

```python
print(dicplayer.keys)
dfplayers = pd.DataFrame.from_dict(dicplayer)

dfplayers.drop(["matches","age"], axis=1, inplace=True)  


filename = "Users/hajiaga/Desktop/DataScience/players.csv"
import os

dfplayers['nationality'] = dfplayers['nationality'].str[3:]

dfplayers["team_county"] = dfplayers['team'].str[:2]

dfplayers["team"] = dfplayers['team'].str[2:]
if os.path.exists(filename):
    os.remove(filename)
dfplayers.to_csv(filename, index=False) 
dfplayers
```
