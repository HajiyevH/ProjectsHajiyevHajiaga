```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Users/hajiaga/Desktop/DataScience/tapazdata.csv")
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
dfs = df.groupby("comp_name")["price"].mean().reset_index()


plt.figure(figsize=(10, 6))
sns.barplot(x='price', y='comp_name', data=dfs, palette='mako')
plt.xlabel('Average Price')
plt.ylabel('Company Name')
plt.title('Average Price for Each Company')
plt.show()
```
